```python
from datetime import datetime
import time
import os
from tqdm import tqdm
import pandas as pd
import re
import matplotlib.pyplot as plt

```

```python
multiple_path = './Modify multiple posts that have been reused multiple times.txt'
multiple_post = []
with open(multiple_path, 'r') as f:
    multiple_post = f.readlines()
print(len(multiple_post))
multiple_post[:5]
```

```python
so_dict = pd.read_pickle('../So Post idx mapping.pkl')
```

```python
so_path = 'D:/CCFinder/So Posts/'
# so_list = os.listdir(so_path)

multiple_lens = []
for k in tqdm(multiple_post):
    idx = so_dict[k]

    with open(so_path+str(idx)+'.txt', 'r', encoding='utf-8') as f:
        lines = f.readlines()[1:]
        codes = ' '.join(lines)
        len_code = len(codes.split(' '))
        multiple_lens.append(len_code)
print(len(multiple_lens))
```

```python
# multiple_lens
```

```python
all_lens = []
for idx in tqdm(range(1355616)):
    with open(so_path+str(idx)+'.txt', 'r', encoding='utf-8') as f:
        lines = f.readlines()[1:]
        codes = ' '.join(lines)
        len_code = len(codes.split(' '))
        all_lens.append(len_code)
print(len(all_lens))
```

```python

code_len_list = [multiple_lens, all_lens]
```

```python
plt.figure(dpi=200)

plt.boxplot(code_len_list,
            labels = ['Multiple modification', 'All'],
            patch_artist=True,
            showmeans=True,
            boxprops = {'color':'black','facecolor':'#1E90FF'},
            showfliers = False,
            meanprops = {'marker':'o','markerfacecolor':'red'},
            medianprops = {'linestyle':'--','color':'black'},) 
plt.grid()
plt.ylabel('The code snippet length')
# plt.yticks(range(2008, 2021), range(2008, 2021))
# plt.xticks(rotation=45)
plt.savefig('./pic/code_len.jpg')
plt.show()
```
