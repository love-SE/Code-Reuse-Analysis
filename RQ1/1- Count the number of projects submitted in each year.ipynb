```python
import pandas as pd
from datetime import datetime
import time
import os
from tqdm import tqdm
import pandas as pd
import re
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('./code reuse ratio in each year of the project.xlsx')
df.info()
```

```python
prj_nums_per_year = []
for i in range(2008, 2021):
    prj_nums_per_year.append(len(df[str(i)].dropna()))
```

```python
prj_nums_per_year
```

```python
# data
import matplotlib.pyplot as plt
x = [i for i in range(len(prj_nums_per_year))]
y = prj_nums_per_year
plt.figure(dpi=300)
# draw

plt.bar(x,y,align="center",color="#1E90FF",tick_label=[i for i in range(2008, 2021)],alpha=1.0)
for a,b,i in zip(x,y,range(len(x))):
    plt.text(a,b+0.3,str(y[i]),ha='center',fontsize=5)
plt.xlabel("Time")
plt.ylabel("Number")
plt.xticks(rotation=45)
plt.grid()
# plt.savefig('./pic/prj_nums_per_year.jpg',bbox_inches = 'tight')
```
