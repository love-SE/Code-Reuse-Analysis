```python
from datetime import datetime
import time
import os
from tqdm import tqdm
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
path_dict = {0:'./1/', 1:'./2/', 2:'./3-5/',
             3:'./6-15/', 4:'./16-100/', 5:'./100/'}
```

```python
df_list = []
for i in range(6):
    df = pd.read_excel(path_dict[i]+'so details.xlsx')
    print(df.info())
    df_list.append(df)
```

```python
df1, df2, df3, df4, df5, df6 = df_list
```

```python
Score1, Score2, Score3, Score4, Score5, Score6 = df1['Score'].dropna().to_list(),\
df2['Score'].dropna().to_list(),\
df3['Score'].dropna().to_list(),\
df4['Score'].dropna().to_list(),\
df5['Score'].dropna().to_list(),\
df6['Score'].dropna().to_list()
Score = [Score1, Score2, Score3, Score4, Score5, Score6]
```

```python
# df1['FavoriteCount'].dropna().to_list()
```

```python
len(Score1),len(Score2),len(Score3),len(Score4),len(Score5),len(Score6)
```

```python
plt.figure(dpi=200)

plt.boxplot(Score,
            labels = [str(i) for i in range(1,7)],
            patch_artist=True,
            showmeans=True,
            boxprops = {'color':'black','facecolor':'#1E90FF'},
            showfliers = False,
            meanprops = {'marker':'o','markerfacecolor':'red'},
            medianprops = {'linestyle':'--','color':'black'},)
plt.grid()
plt.ylabel('Code Reuse Ratio (unit: %)')
plt.xticks(rotation=45)
plt.savefig('./pic/Score.jpg')
plt.show()
```
