```python
from datetime import datetime
import time
import os
from tqdm import tqdm
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python

path_dict = {0:'./1/', 1:'./2/', 2:'./3-5/',
             3:'./6-15/', 4:'./16-100/', 5:'./100/'}
```

```python
df_list = []
for i in range(6):
    df = pd.read_excel(path_dict[i]+'so details.xlsx')
    print(df.info())
    df_list.append(df)
```

```python
df1, df2, df3, df4, df5, df6 = df_list
```

```python
for tf in df_list:
    tf['CreationDate'] = tf['CreationDate'].apply(lambda x:x[:4])
```

```python
df1['CreationDate'].value_counts()
```

```python
time1 = df1['CreationDate'].value_counts()
t1 = list(time1.index)
t1.sort()
t1
# data
import matplotlib.pyplot as plt
x = [i for i in range(len(time1))]
y = [time1.loc[str(i)] for i in t1]
plt.figure(dpi=300)
# draw

plt.bar(x,y,align="center",color="#1E90FF",tick_label=[i for i in t1],alpha=1.0)
for a,b,i in zip(x,y,range(len(x))):
    plt.text(a,b+0.3,str(y[i]),ha='center',fontsize=5)
plt.xlabel("Time")
plt.ylabel("Number")
plt.xticks(rotation=45)
plt.grid()
plt.savefig('./pic/time(1).jpg')
```
