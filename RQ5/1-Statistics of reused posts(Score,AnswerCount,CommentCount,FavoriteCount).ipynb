```python
from datetime import datetime
import time
import os
from tqdm import tqdm
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
bug_df = pd.read_csv('./reuse post/so details.txt')
bug_df.info()
bug_df.head(2)
```

```python
non_bug_df = pd.read_csv('./all post/so details.txt')
non_bug_df.info()
non_bug_df.head(2)
```

```python
Score1, Score2 = bug_df['Score'].dropna().to_list(), non_bug_df['Score'].dropna().to_list()

Score = [Score1, Score2]
```

```python
# df1['FavoriteCount'].dropna().to_list()
```

```python
len(Score1),len(Score2)
```

```python
plt.figure(dpi=200)
plt.boxplot(Score,
            labels = [str(i) for i in range(1,3)],
            patch_artist=True,
            showmeans=True,
            boxprops = {'color':'black','facecolor':'#1E90FF'},
            showfliers = False,
            meanprops = {'marker':'o','markerfacecolor':'red'},
            medianprops = {'linestyle':'--','color':'black'},)
plt.grid()
plt.ylabel('Score')
# plt.yticks(range(2008, 2021), range(2008, 2021))
plt.xticks(rotation=45)
plt.savefig('./pic/Score.jpg')
plt.show()
```
